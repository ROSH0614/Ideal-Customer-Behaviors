Creating new feature for total accepted offer in all campaigns

```python
df.drop(['mnt_wines', 'mnt_fruits','mnt_meat_products', 'mnt_fish_products', 'mnt_sweet_products','mnt_gold_prods'],axis=1,inplace=True)
```

Creating new feature for total number of purchase

```python
df['Total_purchase']=df['num_web_purchases'] + df['num_catalog_purchases'] + df['num_store_purchases'] + df['num_deals_purchases']
```

```python
df.drop(['num_web_purchases', 'num_catalog_purchases','num_store_purchases','num_deals_purchases'],axis=1,inplace=True)
```

# Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
```

```python
df.dtypes
```

```python
cat = ['education','marital_status']
for i in df.columns:
    if (df[i].dtypes == "object"):
        cat.append(i)

print(cat)
```

```python
for x in cat:
    df[x]=df[[x]].apply(le.fit_transform)
```

```python
from sklearn.preprocessing import StandardScaler

ss=StandardScaler()
```

```python
scaled_features = ss.fit_transform(df.values)
scaled_features_df = pd.DataFrame(scaled_features, index=df.index, columns=df.columns)
```

```python
scaled_features_df.head()
```

# Using K-Means 

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
Elbow_M = KElbowVisualizer(KMeans(), k=10,timings=False,metric='silhouette')
Elbow_M.fit(scaled_features_df)
Elbow_M.show()
plt.show()
```

we use this yellowbrick classifier to get K value, so here k=2 which is optimal in this case

```python
kmeans=KMeans(n_clusters=2,random_state=10)
```

```python
kmeans.fit(scaled_features_df)
```

```python
pred=kmeans.predict(scaled_features_df)
df['Cluster']=pred
```
