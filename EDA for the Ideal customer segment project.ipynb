```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import klib
## Display all the columns of the dataframe

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('marketing_campaign.csv')

## print shape of dataset with rows and columns
print(dataset.shape)
```

```python
dataset
```

```python
dataset['Age'] = (2021-dataset['Year_Birth'])

display(dataset)
```

```python
dataset=dataset.drop(['ID','Year_Birth'], axis=1)
dataset
```

```python
df = pd.DataFrame(dataset)
```

```python
klib.missingval_plot(df) # returns a figure containing information about missing values
```

```python
## 1 -step make the list of features which has missing values
features_with_na=[features for features in dataset.columns if df[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values

for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean(), 4),  ' % missing values')
```

```python
# Imputation of mean in Income feature according to Education
df['Income'] = df['Income'].fillna(df.groupby('Education')['Income'].transform('mean'))
```
